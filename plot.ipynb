```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
days=[] # blank array
for i in range(1,32):# taking days input from user by using for loop
    days.append(i)  # append the values in days array
    
daily_temperature=[30,31,31,34,33,32,32,34,40,39,38,33,33,34,35,36,34,37,32,38,39,37,33,34,35,45,34,39,30,34,33]
plt.plot(days,daily_temperature,marker="o")
plt.title("daily_temperature for a  month") # will be show the title in plot.
plt.xlabel("days") # days value going to the x axis.
plt.ylabel("temperature") #temperatures value going to the y axis.
plt.show()  #showing the plot
```

As you can se above the picture 30 is the lowest temperature and 45 is the highest temperature in a month.

```python
# bar plot

categories=['Rent','Groceries','Utilitiees','Entertainment','Transportation']
expenses=[1200,500,2000,339,2500]

plt.figure(figsize=(12,8))
plt.bar(categories,expenses, color="pink")# plot will show pink color.
plt.title("How many money is expenses in a month for which purpose")# will be show the title in plot.
plt.xlabel("categories")#categories value going to the x axis.
plt.ylabel("expenses")#expenses value going to the y axis.
plt.show()#showing the plot.
```

As you can see above the chart for Entertainment cost or expenses is low and for the Transportation express heighest .
